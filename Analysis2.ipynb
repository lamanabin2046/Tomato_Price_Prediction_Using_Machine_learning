1. Linear Regression (LR)

```python
import pandas as pd
import numpy as np

# Load the dataset
# df = pd.read_csv("TOMATO_FINAL.csv")

# # --- 1. Feature Engineering and Preprocessing ---

# # Convert 'Date' to datetime and set as index
# df['Date'] = pd.to_datetime(df['Date'])
# df = df.set_index('Date').sort_index()

# Drop non-numeric/redundant columns
df = df.drop(columns=['Commodity', 'Unit', 'Season'], errors='ignore')

# Convert boolean/dummy columns to integer (0 or 1)
bool_cols = df.select_dtypes(include=['bool']).columns
for col in bool_cols:
    df[col] = df[col].astype(int)

# Create Lag Features (e.g., previous day, week, and month price)
def create_lag_features(data, target_col, lags):
    """Creates lagged features for a specified column."""
    for lag in lags:
        data[f'{target_col}_lag_{lag}'] = data[target_col].shift(lag)
    return data

lags_to_use = [1, 7, 30]
df = create_lag_features(df, 'Average_Price', lags_to_use)

# --- CORRECTED LINES FOR DATETIME INDEX ---
# Explicitly convert the index to DatetimeIndex to ensure attributes are available
df.index = pd.to_datetime(df.index)

# Create Time-based Features
df['DayOfWeek'] = df.index.dayofweek  # This will now work
df['DayOfYear'] = df.index.dayofyear
df['Month'] = df.index.month
# ----------------------------------------

# Drop rows with NaN values created by lagging (the first 30 days)
df = df.dropna()

# --- 2. Defining X and y ---

target = 'Average_Price'
# Features include all columns except the target
features = [col for col in df.columns if col != target]

X = df[features]
y = df[target]

# --- 3. Time-Series Split (80% Train, 20% Test) ---

train_size = int(len(df) * 0.8)

# The split is done sequentially, preserving time order
X_train, X_test = X.iloc[:train_size], X.iloc[train_size:]
y_train, y_test = y.iloc[:train_size], y.iloc[train_size:]

print(f"Total samples: {len(df)}")
print(f"Training samples: {len(X_train)}")
print(f"Testing samples: {len(X_test)}")
```

## Linear Regression

```python
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# A. Initialize and Train the Model
lr_model = LinearRegression()
lr_model.fit(X_train, y_train)

# B. Generate Predictions on the Test Set
lr_pred = lr_model.predict(X_test)

# C. Evaluate Performance (Root Mean Squared Error & R-squared)
lr_rmse = np.sqrt(mean_squared_error(y_test, lr_pred))
lr_r2 = r2_score(y_test, lr_pred)

print(f"--- Linear Regression Performance on Test Set ---")
print(f"RMSE (Error in price): {lr_rmse:.2f}")
print(f"R-squared (Variance explained): {lr_r2:.4f}")
```

```python
import matplotlib.pyplot as plt

# Create a DataFrame for easy plotting comparison
results_df = pd.DataFrame({
    'Actual Price': y_test,
    'Predicted Price': lr_pred
}, index=y_test.index)

# Plotting the comparison
plt.figure(figsize=(14, 6))
plt.plot(results_df.index, results_df['Actual Price'], label='Actual Price', color='blue', linewidth=2)
plt.plot(results_df.index, results_df['Predicted Price'], label='Predicted Price (LR)', color='red', linestyle='--', linewidth=2)
plt.title('Tomato Price Prediction vs. Actuals (Test Set)')
plt.xlabel('Date')
plt.ylabel('Average Price')
plt.legend()
plt.grid(True)
plt.show()
```

```python
# Assuming X, y, df, lr_model, and features list are defined from previous steps.

DAYS_TO_FORECAST = 7
lags_to_use = [1, 7, 30] # The lags used in your feature engineering

# --- 1. Setup Initial Data for Forecast ---

# Get the last actual date from the full dataset
last_date = df.index[-1]
future_dates = pd.date_range(start=last_date + pd.Timedelta(days=1), periods=DAYS_TO_FORECAST, freq='D')

# Get the last 30 actual prices to seed the lag features for the forecast
# This list will be updated in the loop
last_prices = df['Average_Price'].iloc[-30:].tolist()

# Get the last known values for exogenous (non-lag) features
# These are assumed constant (forward-filled) for the forecast days.
base_feature_cols = [col for col in X.columns if not col.startswith('Average_Price_lag_')]
base_features = X.iloc[-1][base_feature_cols]

lr_predictions = []

# --- 2. Recursive Forecasting Loop ---

for i in range(DAYS_TO_FORECAST):
    current_date = future_dates[i]

    # A. Construct Features for the Future Day
    
    # Start with base (exogenous) features
    future_row = base_features.to_dict()

    # Update time-based features for the new date
    future_row['DayOfWeek'] = current_date.dayofweek
    future_row['DayOfYear'] = current_date.dayofyear
    future_row['Month'] = current_date.month

    # Update Lag features using the 'last_prices' list (which contains actuals + previous predictions)
    future_row['Average_Price_lag_1'] = last_prices[-1]
    future_row['Average_Price_lag_7'] = last_prices[-7]
    future_row['Average_Price_lag_30'] = last_prices[-30]
    
    # Convert to DataFrame and ensure feature column order matches training data (CRUCIAL)
    X_future_day = pd.DataFrame([future_row], index=[current_date])
    X_future_day = X_future_day.reindex(columns=features)

    # B. Make Prediction
    lr_pred = lr_model.predict(X_future_day)[0]
    lr_predictions.append(lr_pred)

    # C. Update Lag Prices for the Next Day
    # Use the new prediction as the new 'actual' price for the next day's lag calculation
    last_prices.append(lr_pred)
    last_prices.pop(0) # Remove the oldest price

# --- 3. Final Results Presentation ---

forecast_df_lr = pd.DataFrame({
    'Date': future_dates,
    'LR_Predicted_Price': lr_predictions
})
forecast_df_lr.set_index('Date', inplace=True)

print("--- 7-Day Forecast (Linear Regression) ---")
print(forecast_df_lr.to_markdown())
```

```python
from sklearn.model_selection import TimeSeriesSplit
from sklearn.metrics import r2_score

tscv = TimeSeriesSplit(n_splits=5)
scores = []

for train_idx, test_idx in tscv.split(X):
    X_train, X_test = X.iloc[train_idx], X.iloc[test_idx]
    y_train, y_test = y.iloc[train_idx], y.iloc[test_idx]
    
    lr_model.fit(X_train, y_train)
    pred = lr_model.predict(X_test)
    scores.append(r2_score(y_test, pred))

print(f"Time Series CV R²: {np.mean(scores):.3f}")
```

```python
import pandas as pd
import numpy as np

# Assuming you already have:
# df          -> your full dataset
# best_xgb    -> your trained tuned XGBoost model
# features    -> list of feature columns used for training

# 1️⃣ Get the last known data row
last_row = df.iloc[-1:].copy()

# 2️⃣ Create a container for predictions
future_predictions = []

# 3️⃣ Predict next 7 days iteratively
for i in range(7):
    # Predict next day price
    next_price = best_xgb.predict(last_row[features])[0]

    # Store prediction
    future_predictions.append({
        'Date': last_row.index[-1] + pd.Timedelta(days=1),
        'Predicted_Price': next_price
    })

    # Update lag features for next iteration
    new_row = last_row.copy()
    new_row.index = [last_row.index[-1] + pd.Timedelta(days=1)]
    new_row['Average_Price_lag_1'] = next_price
    new_row['Average_Price_lag_7'] = last_row['Average_Price_lag_6'] if 'Average_Price_lag_6' in last_row else next_price
    new_row['Average_Price_lag_30'] = last_row['Average_Price_lag_29'] if 'Average_Price_lag_29' in last_row else next_price
    
    # Shift other lag features if you have them
    for lag in [2,3,4,5,6]:
        if f'Average_Price_lag_{lag}' in last_row.columns:
            new_row[f'Average_Price_lag_{lag}'] = last_row[f'Average_Price_lag_{lag-1}']

    # Update time-based features (important!)
    new_row['DayOfWeek'] = new_row.index.dayofweek
    new_row['DayOfYear'] = new_row.index.dayofyear
    new_row['Month'] = new_row.index.month

    # Append for next iteration
    last_row = new_row.copy()

# 4️⃣ Convert predictions to DataFrame
future_df = pd.DataFrame(future_predictions)
future_df.set_index('Date', inplace=True)

print("✅ Next 7 Days Predicted Tomato Prices (Kalimati Market)")
print(future_df)
```
