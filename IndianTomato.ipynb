```python
# Rename Nepali columns to English
df = df.rename(columns={
    '‡§ï‡•É‡§∑‡§ø ‡§â‡§™‡§ú': 'commodity',
    '‡§à‡§ï‡§æ‡§á': 'unit',
    '‡§®‡•ç‡§Ø‡•Ç‡§®‡§§‡§Æ': 'minimum',
    '‡§Ö‡§ß‡§ø‡§ï‡§§‡§Æ': 'maximum',
    '‡§î‡§∏‡§§': 'average_price'
})

# Optional: reorder columns to ensure correct order
df = df[['Date', 'commodity', 'unit', 'minimum', 'maximum', 'average_price']]

# Verify
df.info()
df.head()
```

```python
df['Date'] = pd.to_datetime(df['Date'], format='mixed', errors='coerce')
```

```python
df.head
```

```python
Indian_Tomato = df[df['commodity']=='‡§ó‡•ã‡§≤‡§≠‡•á‡§°‡§æ ‡§†‡•Ç‡§≤‡•ã(‡§≠‡§æ‡§∞‡§§‡•Ä‡§Ø)']
```

```python
Indian_Tomato.head()
```

```python
Indian_Tomato.tail()
```

```python
# üß© 1Ô∏è‚É£ Extract numeric parts from Nepali prices (removing ‡§∞‡•Ç and spaces)
Indian_Tomato['minimum'] = Indian_Tomato['minimum'].str.extract(r'(\d+\.?\d*)')
Indian_Tomato['maximum'] = Indian_Tomato['maximum'].str.extract(r'(\d+\.?\d*)')
Indian_Tomato['average_price'] = Indian_Tomato['average_price'].str.extract(r'(\d+\.?\d*)')

# üß© 2Ô∏è‚É£ Convert Nepali digits ‚Üí English digits
nepali_to_english = str.maketrans('‡•ß‡•®‡•©‡•™‡•´‡•¨‡•≠‡•Æ‡•Ø‡•¶', '1234567890')

Indian_Tomato[['minimum', 'maximum', 'average_price']] = (
    Indian_Tomato[['minimum', 'maximum', 'average_price']]
    .apply(lambda col: col.astype(str).map(lambda x: x.translate(nepali_to_english)))
)

# üß© 3Ô∏è‚É£ Convert to numeric type (floats)
Indian_Tomato[['minimum', 'maximum', 'average_price']] = Indian_Tomato[['minimum', 'maximum', 'average_price']].apply(pd.to_numeric, errors='coerce')

# üß© 4Ô∏è‚É£ (Optional) Verify result
print(Indian_Tomato.head())
print(Indian_Tomato.dtypes)
```
