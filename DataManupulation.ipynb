```python
# Rename Nepali columns to English
df2 = df2.rename(columns={
    'рдХреГрд╖рд┐ рдЙрдкрдЬ': 'commodity',
    'рдИрдХрд╛рдЗ': 'unit',
    'рдиреНрдпреВрдирддрдо': 'minimum',
    'рдЕрдзрд┐рдХрддрдо': 'maximum',
    'рдФрд╕рдд': 'average_price'
})

# Optional: reorder columns to ensure correct order
df2 = df2[['Date', 'commodity', 'unit', 'minimum', 'maximum', 'average_price']]

# Verify
df2.info()
df2.head()
```

```python
tomato = df[df['commodity']=='рдЧреЛрд▓рднреЗрдбрд╛ рдареВрд▓реЛ(рдиреЗрдкрд╛рд▓реА)']
```

```python
tomato.head()
```

```python
# ЁЯзй 1я╕ПтГг Extract numeric parts from Nepali prices (removing рд░реВ and spaces)
tomato['minimum'] = tomato['minimum'].str.extract(r'(\d+\.?\d*)')
tomato['maximum'] = tomato['maximum'].str.extract(r'(\d+\.?\d*)')
tomato['average_price'] = tomato['average_price'].str.extract(r'(\d+\.?\d*)')

# ЁЯзй 2я╕ПтГг Convert Nepali digits тЖТ English digits
nepali_to_english = str.maketrans('резреирейрекрелремренреорепреж', '1234567890')

tomato[['minimum', 'maximum', 'average_price']] = (
    tomato[['minimum', 'maximum', 'average_price']]
    .apply(lambda col: col.astype(str).map(lambda x: x.translate(nepali_to_english)))
)

# ЁЯзй 3я╕ПтГг Convert to numeric type (floats)
tomato[['minimum', 'maximum', 'average_price']] = tomato[['minimum', 'maximum', 'average_price']].apply(pd.to_numeric, errors='coerce')

# ЁЯзй 4я╕ПтГг (Optional) Verify result
print(tomato.head())
print(tomato.dtypes)
```
